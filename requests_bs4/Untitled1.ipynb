```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get('http://www.ale.am.gov.br/tipos-publicacoes/prestacao-contas/').text
s = BeautifulSoup(r, 'lxml')
```

```python
article = s.find('div', class_='clearfix transparencia')
print(article.prettify())
```

```python
help(article.find)
```

```python
select_anos = article.find('select', attrs={'ng-model':'query.ano'})
for ano in select_anos.find_all('option'):
    print(ano.text)
```

```python
select_meses = article.find('select', attrs={'ng-model':'query.mes'})
for mes in select_meses.find_all('option'):
    print(mes.text)
```

```python
select_deps = article.find('select', attrs={'ng-model':'query.dep'})
for dep in select_deps.find_all('option'):
    print(dep.text)
```

```python
teste = article.find('form')
print(teste.prettify())
```
